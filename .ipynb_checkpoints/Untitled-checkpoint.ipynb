```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
PAR = np.load('5000_pen02_PAR_list.npy')
```

```python
PAR
```

```python
def get_agg(obj, sum_over):
    obj_sum = 0
    obj_list = []
    sum_over = sum_over
    for ii, item in enumerate(obj):
        obj_sum += item
        if ii % sum_over == 0:
            obj_list.append(obj_sum / sum_over)
            obj_sum = 0
    obj_agg = np.array(obj_list)
    return obj_agg
```

```python
obj_agg = get_agg(PAR, 50)
```

```python
plt.plot(obj_agg[1:]);
```

```python
reward = np.load('5000_pen02_reward_list.npy')
```

```python
reward
```

```python
obj_agg = get_agg(reward, 50)
```

```python
plt.plot(obj_agg[1:]);
```

```python
Q = np.load('11000_pen0.1_proper_char_Q.npy')
```

```python
l_2 = []
for row in Q:
    l = []
    for column in row:
        max_action = np.argmax(column[0])
        l.append(max_action)
    l_2.append(np.array(l))
l_Q1_0 = np.array(l_2)
```

```python
l_2 = []
for row in Q:
    l = []
    for column in row:
        max_action = np.argmax(column[1])
        l.append(max_action)
    l_2.append(np.array(l))
l_Q1_1 = np.array(l_2)
```

```python
l_2 = []
for row in Q:
    l = []
    for column in row:
        max_action = np.argmax(column[2])
        l.append(max_action)
    l_2.append(np.array(l))
l_Q1_2 = np.array(l_2)
```
